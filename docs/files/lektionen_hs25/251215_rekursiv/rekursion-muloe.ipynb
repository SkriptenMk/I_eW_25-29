```python
"""Aufgabe: finden Sie 2 Fehler"""
def anstossen(n: int) -> int:
    ## Basisfälle
    if n <= 1:
        return 0
    if n == 2:
        return 1  # Korrigiert: Bei 2 Personen wird nur 1 Mal angestossen

    return anstossen(n-1) + (n - 1) # Korrigiert: Man addiert n-1
```

```python
"""Bonusaufgabe: Gauss Rekursiv"""
def gauss_recursive(n : int) -> int:
    if n < 2: # 1-> 1 0->0
        return n
    return gauss_recursive(n-1) + n
    
print(f'Die Summe der Zahlen von 1 '
      f'bis 100 ist {gauss_recursive(100)}.')
```

```python
"""Aufgabe: Fakultät"""
def factorial_recursive(n : int) -> int:
    if n < 2:
        # 0 -> 1
        # 1 -> 1
        return 1
    return factorial_recursive(n-1) * n
    
print(f'Das Resultat von 5! ist '
      f'{factorial_recursive(5)}.')
```

```python
"""Aufgabe: umdrehen"""
def reverse(s : str) -> str:
    if len(s) == 0 or len(s) == 1:
        return s
    else:
        head = s[0]
        tail = s[1:]
        return reverse(tail) + head
```

```python
"""
Bonusaufgabe
Studieren Sie die Funktion und füllen Sie die Lücken mit den Auslassungspunkten.
"""
MUENZEN = [1000, 200, 100, 50, 20, 10, 5, 2, 1]
def anz_zahlmöglichkeiten(
    value: int,
    largest_index: int=0,
    ) -> int:
    """Rekursive Funktion"""
    if value < 0:
        return 0
    if value == 0:
        return 1
    if largest_index + 1 == len(MUENZEN):
        # Pay with smales coins
        return 1
    # Pay with largest Coin
    ret = anz_zahlmöglichkeiten(
        value=value-MUENZEN[largest_index],
        largest_index=largest_index,
    )

    # Pay without larges coin
    ret += anz_zahlmöglichkeiten(
        value=value,
        largest_index=largest_index+1,
    )
    return ret

print(anz_zahlmöglichkeiten(5))
```
