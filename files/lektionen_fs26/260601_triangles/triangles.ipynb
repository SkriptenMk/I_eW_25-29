### Calculating a Triangle's Area Using Heron's Formula

![General Triangle.](perimeter.svg)

To calculate the area $T$ of a triangle, we use Heron's formula

$$
T = \sqrt{s(s-a)(s-b)(s-c)}
$$

where $s$ denotes the *semiperimeter*. The semiperimeter is calculated
by adding the lengths of the triangle's sides and dividing the sum by
two.

$$
s = \frac{1}{2} \cdot (a + b + c)
$$
