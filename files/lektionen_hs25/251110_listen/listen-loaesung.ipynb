````python
"""Aufgabe: zeichne Chomp"""
from pytamaro.de import *

def zeichne_chomp(chomp: list) -> None:
    zelle = kombiniere(
        rechteck(68, 38, rgb_farbe(153, 103, 69)),
        rechteck(70, 40, rgb_farbe(51, 34, 23)),        
    )
    leere_zelle = rechteck(70, 40, weiss)
    tafel = leere_grafik()
    maximum = chomp[0] # groesse der erste Spalte
    for spalte_len in chomp:
        spalte = leere_grafik()
        for i in range(maximum):
            if i < spalte_len:
                spalte = ueber(spalte, zelle)
            else:
                spalte = ueber(spalte, leere_zelle)
        tafel = neben(tafel, spalte)
    zeige_grafik(tafel)
````

````python
# Test von der Aufgabe
chomp_test: list = [4, 3, 3, 3, 2]
zeichne_chomp(chomp_test)
````

````python
"""Aufgabe: programmiere den Spielzug in Chomp"""
def chomp_spielzug(chomp: list, spielzug:list) -> list:
    for i, size in enumerate(chomp):
        if i < spielzug[0]-1:
            # bleibt links liegen
            continue
        if chomp[i] < spielzug[1]:
            # bleibt oben liegen. Es wird nichts mehr abgebochen
            break # `continue` würde auch funktionieren
        chomp[i] = spielzug[1]-1
    return chomp
# Hier können Sie den Code testen:
zeichne_chomp(
    chomp_spielzug([4, 3, 3, 3, 2], spielzug=[3, 2])
)
assert chomp_spielzug([4, 3, 3, 3, 2], spielzug=[3, 2]) == [4, 3, 1, 1, 1]
````

````python
spiel: list = [4, 4, 4, 4, 4, 4]
spiel = chomp_spielzug(spiel, [3, 3])
zeichne_chomp(spiel)
spiel = chomp_spielzug(spiel, [1, 4])
zeichne_chomp(spiel)
spiel = chomp_spielzug(spiel, [4, 2])
zeichne_chomp(spiel)
spiel = chomp_spielzug(spiel, [3, 1])
zeichne_chomp(spiel)
spiel = chomp_spielzug(spiel, [2, 2])
zeichne_chomp(spiel)
spiel = chomp_spielzug(spiel, [2, 1])
zeichne_chomp(spiel)
spiel = chomp_spielzug(spiel, [1, 2])
zeichne_chomp(spiel)
spiel = chomp_spielzug(spiel, [1, 1])
````
