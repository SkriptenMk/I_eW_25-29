## Ergänzen und entfernen

Mit der Methode `append` kann ein Element am Ende der Liste hinzufügen und mit `pop` wieder entfernen.

````python
tasks = ["einkaufen", "spazieren gehen", "staubsaugen"]
tasks.append("essen")
print(f"Neu ist auch 'essen' auf der Taskliste: {tasks}")

tasks.pop(1)
print(f"Welcher Task ist nun erledigt? {tasks}")

tasks.pop()
print(f"Wird kein Index angegebn, so wird das letzte Element in der Liste entfernt: {tasks}")
````
