### Aufgabe: Schreiben Sie ein dict

Schreiben Sie ein `dict` für Ihre Klasse, sodass der Nachname dem Schlüssel/key und der Vorname dem Wert/value entspricht.

```python
"""Aufgabe: schreiben Sie ein dict"""
# Da diese Lösungen öffentlich sind, nehme ich hier berühmte Informatiker,
# anstatt Ihre Namen.

klasse = {
    "Berners-Lee": "Tim",
    "Brandenburg": "Karlheinz",
    "Dahl": "Ole-Johan",
    "Dijkstra": "Edsger W.",
    "Knuth": "Donald E.",
    "Neumann": "John von",
    "Ritchie": "Dennis",
    "Stallman": "Richard",
    "Thompson": "Ken",
    "Tomlinson": "Ray",
    "Torvalds": "Linus",
    "Turing": "Alan",
    "Zuse": "Konrad",
}
```

```python
"""Aufgabe: Schleife"""
for nachname, vorname in klasse.items() :
    print(f"Hallo {vorname} {nachname}")
```

```python
"""Aufgabe: Selektion"""

for nachname, vorname in klasse.items():
    if 'e' in nachname:
        print(nachname)
```

```python
"""Beispiel von einem dict"""
telefonbuch = {
    "Feuerwehr": 118,
    "Polizei": 117,
    "Ambulanz": 144,
    "Rega": 1414,
    "Beratung + Hilfe 147": 147,
    "allgemeiner Notruf": 112,
}
```

```python
"""Aufgabe: Telefonliste"""

threedigtslist = []
for name, nr in telefonbuch.items():
    if nr >= 100 and nr < 1000:
        threedigtslist.append(name)

# Test
assert type(threedigtslist) == list
assert len(threedigtslist) == 5
for soultion in ['Feuerwehr', 'Polizei', 'Ambulanz', 'Beratung + Hilfe 147', 'allgemeiner Notruf']:
    assert soultion in threedigtslist
```
